```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
voltage_data = pd.read_csv('NITK_MainBus_EXP_VOLTAGES.csv')
current_data = pd.read_csv('NITK_MainBus_EXP_CURRENTS.csv')
power_data = pd.read_csv('NITK_MainBus_EXP_POWERS.csv')
losses_data = pd.read_csv('NITK_MainBus_EXP_LOSSES.csv')
```

```python
voltage_data.columns = voltage_data.columns.str.strip()
current_data.columns = current_data.columns.str.strip()
power_data.columns = power_data.columns.str.strip()
losses_data.columns = losses_data.columns.str.strip()
```

```python
voltage_data
```

```python
# Create the plot
plt.figure(figsize=(12, 8))

# Plot pu values for Node1, Node2, and Node3 across different buses - but for simplicity we will plot it only for node 2
# plt.plot(voltage_data.index, voltage_data['pu1'], label='Node1 (pu)', color='blue', marker='o')
plt.plot(voltage_data.index, voltage_data['pu2'], label='Node2 (pu)', color='red', marker='o')
# plt.plot(voltage_data.index, voltage_data['pu3'], label='Node3 (pu)', color='green', marker='o')

# Annotate each point with its pu value
for i in voltage_data.index:
    plt.text(i, voltage_data['pu1'][i], f"{voltage_data['pu1'][i]:.5f}", color='blue', fontsize=12, ha='center')
    plt.text(i, voltage_data['pu2'][i], f"{voltage_data['pu2'][i]:.5f}", color='red', fontsize=12, ha='center')
    plt.text(i, voltage_data['pu3'][i], f"{voltage_data['pu3'][i]:.5f}", color='green', fontsize=12, ha='center')

# Customize the plot
plt.title('Voltage Magnitudes (pu) at Different Nodes Across Buses', fontsize=14)
plt.xlabel('Bus Index', fontsize=12)
plt.ylabel('Magnitude (pu)', fontsize=12)
plt.xticks(ticks=voltage_data.index, labels=voltage_data['Bus'], rotation=0, fontsize=10)
plt.yticks(fontsize=10)
# plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
current_data
```

```python
# Plot the current magnitudes for the first phase (I1_1) across all elements
plt.figure(figsize=(12, 8))

# Plot the current magnitudes
plt.plot(current_data['Element'], current_data['I1_1'], label='Phase 1 Current Magnitude (I1_1)', color='blue', marker='o')

# Annotate each point with its magnitude value
for i in range(len(current_data)):
    plt.text(i, current_data['I1_1'][i], f"{current_data['I1_1'][i]:.2f}", color='black', fontsize=9, ha='center')

# Customize the plot
plt.title('Phase 1 Current Magnitudes Across Elements', fontsize=14)
plt.xlabel('Element', fontsize=12)
plt.ylabel('Current Magnitude (A)', fontsize=12)
plt.xticks(ticks=range(len(current_data)), labels=current_data['Element'], rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
power_data
```

```python
terminal_1_data = power_data[power_data['Terminal'] == 1]

# Create the plot
plt.figure(figsize=(12, 8))

# Plot Active Power (P(kW)) for Terminal 1
plt.plot(terminal_1_data['Element'], terminal_1_data['P(kW)'], label='Active Power (P)', color='blue', marker='o')

# Plot Reactive Power (Q(kvar)) for Terminal 1
plt.plot(terminal_1_data['Element'], terminal_1_data['Q(kvar)'], label='Reactive Power (Q)', color='red', marker='o')

# Annotate each point with its value
for i in range(len(terminal_1_data)):
    plt.text(i, terminal_1_data['P(kW)'].iloc[i], f"{terminal_1_data['P(kW)'].iloc[i]:.1f}", color='blue', fontsize=9, ha='center')
    plt.text(i, terminal_1_data['Q(kvar)'].iloc[i], f"{terminal_1_data['Q(kvar)'].iloc[i]:.1f}", color='red', fontsize=9, ha='center')

# Customize the plot
plt.title('Active and Reactive Power for Terminal 1', fontsize=14)
plt.xlabel('Element', fontsize=12)
plt.ylabel('Power (kW/kvar)', fontsize=12)
plt.xticks(ticks=range(len(terminal_1_data)), labels=terminal_1_data['Element'], rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
losses_data
```

```python
# Create the plot
plt.figure(figsize=(12, 8))

# Plot I2R(W)
plt.plot(losses_data['Element'], losses_data['I2R(W)'], label='I2R Losses (W)', color='blue', marker='o')

# Plot I2X(var)
plt.plot(losses_data['Element'], losses_data['I2X(var)'], label='I2X Losses (var)', color='red', marker='o')

# Annotate each point with its value
for i in range(len(losses_data)):
    plt.text(i, losses_data['I2R(W)'][i], f"{losses_data['I2R(W)'][i]:.2f}", color='black', fontsize=9, ha='center')
    plt.text(i, losses_data['I2X(var)'][i], f"{losses_data['I2X(var)'][i]:.2f}", color='black', fontsize=9, ha='center')

# Customize the plot
plt.title('I2R and I2X Losses Across Elements', fontsize=14)
plt.xlabel('Element', fontsize=12)
plt.ylabel('Losses (W/var)', fontsize=12)
plt.xticks(ticks=range(len(losses_data)), labels=losses_data['Element'], rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
